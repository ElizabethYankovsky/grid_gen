Start with a great arc passing from $B$ to $C$ and know that $ABCD$ will form a box with $AB$ and $CD$ also great arcs and $ABC$ and $BCD$ being right angles.

- What curve is $AD$ is perpendicular to $AB$ and $CD$?
- What are the scale factors and grid metrics for a grid laid down within $ABCD$?

Consider the following dequence of rotations that map the three known sides of the box (great arcs, $AB$, $BC$, and $CD$) to a regular box in a familiar orthogonal coordinate system.

1. What rotation $R_1$ brings $B$ to the Greenwich Meridian (i.e. longitude of 0$^\circ$E)? (This is for convenience)

  - A rotation about the North pole of $-\lambda_B$.

2. What rotation $R_2$ then brings $R_1 B$ on to the Equator?

  - A rotation of $-\phi_B$ about the pole passing through the Equator at longitudes $\pm 90^\circ$.

3. What rotation about $R_3 B$ then brings $R_2 R_1 C$ on to the Equator so that $R_3 R_2 R_1 AB$ is in the Northern hemisphere?

  - A rotation about the x-axis by some angle that has to be determined.

$B'C'$ is now a great arc on the Equator. Label the resulting box $A'B'C'D' = R_3 R_2 R_1 ABCD$.

- If $A'B'$ is perpendicular to $B'C'$ and is a great arc then it falls on a Meridian.
- If $C'D'$ is perpendicular to $B'C'$ and is a great arc then it falls on a Meridian.

Requiring $AD$ to be perpendicular to both $AB$ and $CD$ is equivalent to requiring $A'D'$ to be perpendicular to both $A'B'$ and $C'D'$ since rotations do not change angles between lines. The curve that is perpendicular to a Meridian is a line of constant latitude. Thus $A'D'$ is a line of constant latitude and $AD$ is an small arc obtained by transforming $R_1^{-1}R_2^{-1}R_3^{-1}$ the line of constant latitude, $A'D'$.

### Grid generation

A grid that fits $A'B'C'D'$ has nodes given by

\begin{align}
\lambda'(i,j) &= i \Delta \lambda \;\; \mbox{for} \;\; i = 0, 1, \ldots, N \\
\phi'(i,j) &= j \Delta \phi \;\; \mbox{for} \;\; j = 0, 1, \ldots, M
\end{align}

## Working in 3D coordinates (x,y,z)

Use the right-hand coordiante system $x,y,z$ where the x-axis ($y=z=0$) passes through the Greenwich Meridian on the equator, y-axis ($x=0,z=0$) is at right angles and on the EQuator and the z-axis ($x=0,y=0$) passes through the North pole.

The position in space of a point with geographical coordinates $\lambda,\phi$ (corresponding to longitude and latitude measured in radians) is

\begin{align}
\begin{pmatrix}
x \\ y \\ z
\end{pmatrix}
&=
\begin{pmatrix}
R_e \cos{\phi} \cos{\lambda} \\
R_e \cos{\phi} \sin{\lambda} \\
R_e \sin{\phi}
\end{pmatrix}
\end{align}

where $R_e$ is the (spherical) radius of the Earth.

The geographic coordiantes of a point $\mathbf{r}=(x,y,z)$ projected to the Earth's surface from the origin is

\begin{align}
\phi &= \sin^{-1} \frac{z}{|\mathbf{r}|} \\
\lambda &= \tan^{-1} \frac{y}{x}
\end{align}

Note that for points on the Earth's surface, $|r|=Re$. The above rotations can now be described in terms of rotations matrices that operate on the position vectors of points we are concerned with.

A rotation about the North pole of angle $\alpha$ is given by

\begin{align}
\begin{pmatrix}
\cos \alpha & -\sin \alpha & 0 \\
\sin \alpha & \cos \alpha & 0 \\
0 & 0 & 1
\end{pmatrix}
\end{align}

so $R_1$ is

\begin{align}
R_1 =
\begin{pmatrix}
\cos \lambda_B & \sin \lambda_B & 0 \\
-\sin \lambda_B & \cos \lambda_B & 0 \\
0 & 0 & 1
\end{pmatrix}
\end{align}

A rotation about the y-axis of angle $\beta$ is given by

\begin{align}
\begin{pmatrix}
\cos \beta & 0 & \sin \beta \\
0 & 1 & 0 \\
-\sin \beta & 0 & \cos \beta
\end{pmatrix}
\end{align}

so $R_2$ is

\begin{align}
R_2 &=
\begin{pmatrix}
\cos \phi_B & 0 & -\sin \phi_B \\
0 & 1 & 0 \\
\sin \phi_B & 0 & \cos \phi_B
\end{pmatrix}
\end{align}

A rotation about the x-axis of angle $\gamma$ is given by

\begin{align}
\begin{pmatrix}
1 & 0 & 0 \\
0 & \cos \gamma & -\sin \gamma \\
0 & \sin \gamma & \cos \gamma
\end{pmatrix}
\end{align}

Now $\gamma$ is determined by the position of $C$ after $R_2 R_1$ which is

\begin{align}
R_2 R_1 C &=
\begin{pmatrix}
\cos \phi_B & 0 & -\sin \phi_B \\
0 & 1 & 0 \\
\sin \phi_B & 0 & \cos \phi_B
\end{pmatrix}
\begin{pmatrix}
\cos \lambda_B & \sin \lambda_B & 0 \\
-\sin \lambda_B & \cos \lambda_B & 0 \\
0 & 0 & 1
\end{pmatrix}
\begin{pmatrix}
R_e \cos{\phi_C} \cos{\lambda_C} \\
R_e \cos{\phi_C} \sin{\lambda_C} \\
R_e \sin{\phi_C}
\end{pmatrix}
\end{align}



The Haversine formula should be used (for accuracy) to calculate the length $B'C'=BC$:

\begin{align}
B'C'/R_e &= BC/R_e \\
&= 2 \sin^{-1} \left( \sqrt{ \sin^2 \left( \frac{\phi_B - \phi_C}{2} \right) +
\cos \phi_B \cos \phi_C \sin^2 \left( \frac{\lambda_B - \lambda_C}{2} \right)}\right)
\end{align}

That angle can be found by dropping a perpendicular (great arc) from $C'$ to the Equator. Label the intersection point $P$. From the spherical law of cosines we then know angle $\gamma_{C'}$ between $B'P$ (along the Equator) and $B'C'$ is

\begin{align}
\cos \gamma_{C'}
&= \frac{\cos PC'/R_e - \cos B'P/R_e \cos B'C'/R_e}{ \sin B'P/R_e \sin B'C'/R_e } \\
&= \frac{\cos \phi_{C'} - \cos \lambda_{C'} \cos B'C'/R_e}{ \sin \lambda_{C'} \sin B'C'/R_e }
\end{align}

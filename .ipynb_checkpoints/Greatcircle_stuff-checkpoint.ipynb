```python
# OBJECTIVES:
#   Get First lat/lon from click
#   Query the resolution from the user
#   Get Second lat/lon from click  
#   Define great circle equation
#      - use to provide forward azimuth 
#      - from starting point, work from 1st point adding distance
#   Question: how does great circle deal with spacing? 
#   Question: how to calculate orthoganol points (along great circles)?
```

```python
from math import sin, cos, sqrt, atan2, radians

# approximate radius of earth in km
R = 6373.0

for nt in range(len(x)-1):
  lat1 = radians(y[nt])
  lon1 = radians(x[nt])
  lat2 = radians(y[nt+1])
  lon2 = radians(x[nt+1])

  dlon = lon2 - lon1
  dlat = lat2 - lat1

  a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
  c = 2 * atan2(sqrt(a), sqrt(1 - a))

  distance = R * c

  print("Result:", distance)
```
